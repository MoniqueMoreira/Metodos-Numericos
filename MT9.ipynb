```python
import matplotlib.pyplot as plt
import math

def Newton(p0,tol,Nmax):
  for i in range(Nmax):
    p = p0 - (f(p0)/f_linha(p0))
    if abs(p - p0) <= tol:
      return (p,f(p))
    else:
      p0 = p
  print("Nmax atingido")

def Plot(a,b,delta,x,y):
  while a <= b:
    x.append(a)
    y.append(f(a))
    a = round(a + delta,2)

def Newton_aprox(p0,tol,Nmax,pn,fpn):
  for i in range(Nmax):
    pn.append(p0)
    fpn.append(f(p0))
    p = p0 - (f(p0)/f_linha(p0))
    if abs(p - p0) <= tol:
      return 0
    else:
      p0 = p
  print("Nmax atingido")
```

```python
def main(a,b,p0,delta,tol,Nmax):
  x = []
  y = []
  pn =[]
  fpn = []

  Plot(a,b,delta,x,y)
  plt.plot(x,y,label = "f(x)")

  p = Newton(p0,tol,Nmax)
  plt.scatter(p[0],p[1], color='red', label = 'raiz')

  Newton_aprox(p0,tol,Nmax, pn,fpn)
  plt.scatter(pn,fpn, label = "Aproximação", color ='green')

  plt.legend()
  plt.show()
```

# **a) x − 2^(−x) = 0, [0, 1]**

```python
def f(x):
  return x - 2**(-x)

def f_linha(x):
  return 1 + math.log(2)*2**(-x)

main(0,1,0,0.1,0.001,1000)
```

# **B) x + 1 − 2 sen(πx)= 0, [0, 0.5]**

```python
def f(x):
  return x + 1 - 2*math.sin(math.pi*x)

def f_linha(x):
  return 1 - 2*math.pi*math.cos(math.pi*x)

main(0,0.5,0,0.01,0.001,1000)
```

# **c) 2x cos(2x) − (x + 1)^2 = 0, [−3, −2]**

```python
def f(x):
  return 2*x*math.cos(2*x) - (x+ 1)**2

def f_linha(x):
  return -2*(1 + x - math.cos(2*x) + 2*x*math.sin(2*x))

main(-3,-2,-2,0.1,0.001,1000)
```

# **d) ln(x) − 2^x + x^2 = 0, [3, 5]**

```python
def f(x):
  return math.log(x) - 2**(x) + x**(2)

def f_linha(x):
  return (1/x) + 2*(x) - math.log(2)*2**(x)

main(3,5,4,0.1,0.001,100000)
```
